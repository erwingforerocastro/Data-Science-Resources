### Preprocesando texto natural

```
text5 = '"Ethics are built right into the ideals and objectives of the United Nations" \
#UNSG @ NY Society for Ethical Culture bit.ly/2guVelr'
text6 = text5.split(' ')

text6
```

```
#buscando hastags
[w for w in text6 if w.startswith('#')]
#buscando llamados
[w for w in text6 if w.startswith('@')]
```

```
text7 = '@UN @UN_Women "Ethics are built right into the ideals and objectives of the United Nations" \
#UNSG @ NY Society for Ethical Culture bit.ly/2guVelr'
text8 = text7.split(' ')
```

<br>

Podemos usar expresiones regulares para ayudarnos con un análisis más complejo.

Por ejemplo, '' @ [A-Za-z0-9 _] + '' devolverá todas las palabras que:
* comience con '' @ '' y siga al menos uno:
* mayúscula (`` A-Z '')
* letra minúscula (`` a-z '')
* número (`` 0-9 '')
* o guión bajo (`` _ '')

```
import re # importamos re - un módulo que proporciona soporte para expresiones regulares

[w for w in text8 if re.search('@[A-Za-z0-9_]+', w)]
```

## datos de texto en pandas

```
import pandas as pd

time_sentences = ["Monday: The doctor's appointment is at 2:45pm.", 
                  "Tuesday: The dentist's appointment is at 11:30 am.",
                  "Wednesday: At 7:00pm, there is a basketball game!",
                  "Thursday: Be back home by 11:15 pm at the latest.",
                  "Friday: Take the train at 08:10 am, arrive at 09:00am."]

df = pd.DataFrame(time_sentences, columns=['text'])
df
```

```
#numero de caracteres para cada string
df['text'].str.len()
```

```
#numero de tokens para cada cadena 
df['text'].str.split().str.len()
```

```
#saber si el texto contiene 'appointment'
df['text'].str.contains('appointment')
```

```
#encuentra cuántas veces aparece un dígito en cada cadena
df['text'].str.count(r'\d')
```

```
#obtiene los digitos de la cadena
df['text'].str.findall(r'\d')
```

```
#agrupar y encontrar las horas y minutos
df['text'].str.findall(r'(\d?\d):(\d\d)')
```

```
#reemplazar los dias con ???
df['text'].str.replace(r'\w+day\b', '???')
```

```
#reemplazar los días de semana con abreviaturas de 3 letras
df['text'].str.replace(r'(\w+day\b)', lambda x: x.groups()[0][:3])
```

```
#crear nuevas columnas desde la primera coincidencia de grupos extraídos
df['text'].str.extract(r'(\d?\d):(\d\d)')
```

```
#extraer todo el tiempo, las horas, los minutos y el período
df['text'].str.extractall(r'((\d?\d):(\d\d) ?([ap]m))')
```

```
# extraer todo el tiempo, las horas, los minutos y el período con los nombres de los grupos
df['text'].str.extractall(r'(?P<time>(?P<hour>\d?\d):(?P<minute>\d\d) ?(?P<period>[ap]m))') #para nombrar ?P<nombre>
```

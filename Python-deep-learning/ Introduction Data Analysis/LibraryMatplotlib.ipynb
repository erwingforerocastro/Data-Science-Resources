```
from matplotlib.artist import Artist

def rec_gc(art, depth=0):
    if isinstance(art, Artist):
        # incrementar la profundidad para cuando se pinte (espaciado)
        print("  " * depth + str(art))
        for child in art.get_children():
            rec_gc(child, depth+2)

# Llame a esta función en el artista de la leyenda para ver de qué está compuesta la leyenda
rec_gc(plt.legend())
```

```
from google.colab import files
import numpy as np
from matplotlib import dates
from datetime import datetime
import pandas as pd

files.upload() #subimos el archivo Ruta del proyecto /External Files/Introduction Data Analisis (necesario permitir las cookies)
df = pd.read_csv('/content/GHCN-Daily.csv')
years={}
for year in range(2005,2016):
    m='{}-01-01'.format(year)
    f='{}-12-31'.format(year)
    current_date=df[(df['Date']>=m) & (df['Date']<=f)]
    year_date=[]
    for date in current_date['Date'].sort_values().unique():
        if '02-29' not in date[-5:]:
            year_date.append([date.replace('-',''),current_date[current_date['Date']==date]['Data_Value'].min(),
                              current_date[current_date['Date']==date]['Data_Value'].max()])
    years[year]=year_date 

max_min=pd.DataFrame({'Min':np.array(np.zeros(365))+150,'Max':np.array(np.zeros(365))},index=np.arange(0,365))
record_2015=pd.DataFrame({'Min':np.array(np.zeros(365)),'Max':np.array(np.zeros(365)),'Date':['']*365},index=np.arange(0,365))
for year in range(2005,2016):
    year_current=pd.DataFrame(years[year],columns=['Date','Min','Max'])
    if year!=2015:
        for day in range(0,365):
            if max_min.loc[day,'Max']<year_current['Max'][day]:
                max_min.loc[day,'Max']=year_current['Max'][day]
                
            if max_min.loc[day,'Min']>year_current['Min'][day]:
                max_min.loc[day,'Min']=year_current['Min'][day]
    else:
        for day in range(0,365):
            if year_current.loc[day,'Max']>max_min['Max'][day]:
                record_2015.loc[day,'Max']=year_current['Max'][day]
                record_2015.loc[day,'Date']=year_current['Date'][day].replace('2015','2005')
            else:
                record_2015.loc[day,'Max']=None
                
            if (year_current.loc[day,'Min']<max_min['Min'][day]):
                record_2015.loc[day,'Min']=year_current['Min'][day]
                record_2015.loc[day,'Date']=year_current['Date'][day].replace('2015','2005')
            else:
                record_2015.loc[day,'Min']=None

y1_2015=record_2015[['Max','Date']].dropna()
x1_2015=y1_2015['Date']
y2_2015=record_2015[['Min','Date']].dropna()
x2_2015=y2_2015['Date']
  
#graficamos los valores
plt.figure(figsize=(20,12))
year_current=pd.DataFrame(years[2005],columns=['Date','Min','Max'])
#convertimos los valores del eje x
xaxis=[datetime.strptime(d, '%Y%m%d') for d in year_current['Date']]
xaxis1=[datetime.strptime(d, '%Y%m%d') for d in x1_2015]
xaxis2=[datetime.strptime(d, '%Y%m%d') for d in x2_2015]
ax = plt.gca()
xaxis = dates.date2num(xaxis) 
xaxis1=dates.date2num(xaxis1)
xaxis2=dates.date2num(xaxis2)
#asignamos el tipo de formato del eje x
hfmt = dates.DateFormatter('%m / %d')
ax.xaxis.set_major_formatter(hfmt)


#graficamos los valores 
plt.scatter(xaxis1,y1_2015['Max'],marker='o',s=200,label='Maximum record temp. 2015')
plt.scatter(xaxis2,y2_2015['Min'],marker='o',s=200,label='Minimum record temp. 2015')
lines=plt.plot(xaxis,max_min['Min'],xaxis,max_min['Max'])
l1, l2 = lines
plt.setp(lines, linestyle='-')
plt.setp(l1,linewidth=2,color='#3346FF',label='Minimum Temp. 2005-2014')
plt.setp(l2,linewidth=2,color='#FF3333',label='Maximum Temp. 2005-2014')
            
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel('Dates [mm/dd]',fontsize=20)
plt.ylabel('temp [C°]',fontsize=20)
plt.title('Source: (GHCN-Daily) ', loc='left')
plt.title('Record temperatures 2005-2014 vs Record temperatures 2015',fontsize=20)
plt.legend(fontsize=15)
plt.gca().fill_between(xaxis,
                       max_min['Min'],max_min['Max'],
                       facecolor='#88FC9F',
                       alpha=0.2)
plt.show()
```

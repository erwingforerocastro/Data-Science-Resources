```
#guardar un archivo en txt
a=np.random.randint(0,10,(3,5))
np.savetxt('./numpy/ejemplo.txt',a) 
#guardar un archivo en csv
np.savetxt('./numpy/ejemplo.csv',a,delimiter=';') 
#guardar un archivo en binario
np.save('./numpy/ejemplo.npy',a) 
#guardar un archivo en binario comprimido
np.savez('./numpy/ejemplo.npz',a) 
```

```
#cargar el archivo guardado
a=np.loadtxt('./numpy/ejemplo.txt') 
b=np.loadtxt('./numpy/ejemplo.csv',delimiter=';')
c=np.load('./numpy/ejemplo.npy')
d=np.load('./numpy/ejemplo.npz')['arr_0']
print(a)
print()
print(b) 
print()
print(c)
print()
print(d)
```

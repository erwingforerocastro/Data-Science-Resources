#Implementaci√≥n de un modelo basico convolucional con keras 



```
#importar librerias
import tensorflow as tf
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Conv2D 
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.utils import to_categorical

model = Sequential() 
model.add(Conv2D(32, (5, 5), activation='relu', input_shape=(28, 28, 1))) #32 filtros, kernel 5x5 imagen 28x28 y stride 1
model.add(MaxPooling2D((2, 2)))                                           #kernel 2x2
model.add(Conv2D(64, (5, 5), activation='relu'))                          #64 filtros, ventana o kernel 5x5
model.add(MaxPooling2D((2, 2)))                                           #kernel 2x2
model.add(Flatten())                                                      #aplanamos la salida 3D a 1D
model.add(Dense(10,activation='softmax'))                                 
model.summary()

#Layer (type)                 Output Shape              Param #   
#=================================================================
#conv2d (Conv2D)              (None, 24, 24, 32)        832         # conv2d da como resultado 24x24, filtros 32            832= (25 neuronas  mas 1 sesgo)x32
#_____________________________________
#max_pooling2d (MaxPooling2D) (None, 12, 12, 32)        0           # max_pooling2d con cada filtro 24x24 da como resultado 12x12, filtros 32
#_________________________________________________________________
#conv2d_1 (Conv2D)            (None, 8, 8, 64)          51264       # conv2d con cada filtro 12x12 da como resultado 8x8, filtros 64   51264= ((25 neuronas x 32 filtros) mas 1 sesgo)x64
#_____________________________________
#max_pooling2d_1 (MaxPooling2 (None, 4, 4, 64)          0           # max_pooling2d_1 con cada filtro 8x8 da como resultado 4x4, filtros 64
#flatten (Flatten)            (None, 1024)              0         
#_________________________________________________________________
#dense (Dense)                (None, 10)                10250       # dense con toda la capa 64x4x4 da como resultado vector de 10 x(1024 mas 1 sesgo)
#=================================================================
```

```
#preparamos loa datos
mnist = tf.keras.datasets.mnist 

(train_images, train_labels), (test_images, test_labels)  = mnist.load_data() 

train_images = train_images.reshape((60000, 28, 28, 1)) #convertimos las imagenes en tensores 3D
train_images = train_images.astype('float32') / 255     #normalizamos

test_images = test_images.reshape((10000, 28, 28, 1)) 
test_images = test_images.astype('float32') / 255 

train_labels = to_categorical(train_labels) #realizamos en one shot
test_labels = to_categorical(test_labels) 
```

```
model.compile(loss='categorical_crossentropy', #funcion objetivo
              optimizer='sgd',                 #sgd gradiente de descenso con momentum
              metrics=['accuracy'])            #metrica precision
```

```
#entrenamiento
model.fit(train_images, train_labels,  batch_size=100,  epochs=5,  verbose=1) #entrenamos el modelo enviamos 100 muestras a sgd entrenamos 5 veces y veremos la barra de progreso
test_loss, test_acc = model.evaluate(test_images, test_labels) 
print('Test precision:', test_acc)  
```

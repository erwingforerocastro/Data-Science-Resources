```
#indexación nos permite traer un valor por su posicion recordar que empieza en 0
x = [5, "deep", 1.8] 
print ("primer elemento --> x[0]: ", x[0]) 
print ("segundo elemento --> x[1]: ", x[1]) 
print ("último elemento --> x[-1]:", x[-1]) 
print ("penúltimo elemento --> x[-2]:", x[-2]
       
print("todos los índices -->") 
print("x[:]: ", x[:]) 
print("índice 1 al final de la lista --> ") 
print("x[1:]: ", x[1:]) 
print("índice 1 al índice 2 (sin incluir el índice 2) --> ") 
print("x[1:2]:", x[1:2]) 
print("índice 0 al último índice (sin incluir el último) --> ") 
print("x[:-1]:", x[:-1])        
```

```
#Tuplas
#Las tuplas son colecciones ordenadas e inmutables 
#(es decir, no se puede modificar su contenido una vez creadas) 
#que se usan para almacenar valores que nunca cambiarán. 

x = (1,"deep",3) 
print (x) 
print (x[1]) 

x = x + (4,"learning") 
print (x) 
print (x[-1]) 
```

```
#Diccionarios 
#Los diccionarios son una colección desordenada, mutable e indexada de pares clave-valor. Es decir, 
#representan arrays asociativos que contienen una clave para asociar cada elemento del diccionario. Existe una limitación, 
#y es que no puede haber claves repetidas en un mismo diccionario, ya que Python sobrescribirá el valor para dicha clave. 

persona = {'nombre': 'Jordi Torres', 'profesion': 'profesor'} 
print (persona) 
print (persona['nombre']) 
print (persona['profesion'])

#Fácilmente se puede cambiar un valor a través de su clave o añadir un nuevo par clave-valor: 
persona['profesion'] = 'investigador' 
print (persona) 

#agregar uno nuevo
persona['grupo'] = 24 
print (persona) 

print (len(persona)) 
```

```
#sentencia if
x = 6.5 
if x < 5:  
  puntuacion = 'baja' 
elif x <= 7: # elif = else if  
  puntuacion = 'media' 
else:  
  puntuacion = 'alta' 

print (puntuacion)  
```

```
#Bucle for
for a in range (1,4):  #range (enumera los miembros de una lista y obtiene una lista de números) 
  print (a) 

universidades = ["UPC", "UPM", "UPV"]  #recorrer un arreglo
print('For con array:')
for universidad in universidades:  
  print (universidad)

print('For breack:')
for universidad in universidades:   #'breack' rompe con el ciclo
   if universidad == "UPM":  
      break  
   print (universidad)  #imprime solo la primera por que al llegar a UPM se rompe y no llega a print

print('For continue:')
for universidad in universidades:  #'continue' pasa al siguiente y no ejecuta lo que sigue de codigo
  if universidad == "UPM":  
    continue                            
  print (universidad) 
```

```
#Bucle While
x = 3 
while x > 0:   #mientras x sea mayor a 0 hara lo siguiente
  x = x - 1  
  print (x)
```

```
#Clases
class Calculadora(object):
     
      def __init__(self): 
          self.valor = 0  

      def suma(self, n):  
          self.valor += n 

      def getValor(self): 
          return self.valor 

calc = Calculadora() 
calc.suma(2) 
print(calc.getValor()) 
calc.suma(2) 
print(calc.getValor()) 
```

```
#Decoradores 

def voy_a_imprimir_Hola():  
  print("Hola") 
def my_decorator(func):  
  def wrapper(): 
    print("[Aviso: voy a decir algo]")  #realizar algo antes
    func()                              #ejecutamos la funcion 
    print("[Aviso: ya he dicho algo]")  #realizar algo despues
  return wrapper 

voy_a_imprimir_Hola = my_decorator(voy_a_imprimir_Hola) #el parametro es el nombre de la funcion, esto crea una funcion anonima
voy_a_imprimir_Hola ()                                  #que ejecutamos

print(" ");
#se puede definir añadiendo @my_decorator como anotación justo antes de la definición de nuestra función: 

@my_decorator                  #nombre de la funcion decoradora
def voy_a_imprimir_Hola():  
  print("Hola")
def my_decorator(func):  
  def wrapper(): 
    print("[Aviso: voy a decir algo]")  #realizar algo antes
    func()                              #ejecutamos la funcion 
    print("[Aviso: ya he dicho algo]")  #realizar algo despues
  return wrapper 

voy_a_imprimir_Hola ()

#Como veremos, los decoradores son usados por TensorFlow 2.0, por ejemplo el decorador @tf.funtion80. Cuando se anota una función con @tf.function
```

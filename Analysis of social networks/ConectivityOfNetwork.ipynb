###Coeficiente de agrupación

```
import networkx as nx
G=nx.Graph()
G.add_edges_from([('A','K'),('A','B'),('A','C'),('B','C'),('B','K'),
                  ('C','E'),('C','F'),('D','E'),('E','F'),('E','H'),('F','G'),('I','J')])
nx.clustering(G,'F')
```

```
#promedio local de agrupación de todos los nodos
nx.average_clustering(G)
```

```
#relación entre el número de triángulos y el número de "tríadas abiertas" en una red
nx.transitivity(G)
```

```
#distancia entre los nodos
nx.shortest_path(G,'A','G')
```

```
#el tamaño de la distancia de los nodos
nx.shortest_path_length(G,'A','G')
```

```
#busqueda de primero en amplitud (arbol)
T=nx.bfs_tree(G,'A')
T.edges() 
```

```
#el promedio del tamaño de la distancia de los nodos
nx.average_shortest_path_length(T)
```

```
#distancia maxima entre dos nodos
G2=nx.Graph()
G2.add_edges_from([('A','B'),('B','C'),('C','F'),('C','E')])
nx.diameter(G2)
```

```
#exentricidad, de un nodo n la distancia entre los otros n nodos
nx.eccentricity(G2)
```

```
#el radio, es el de minima exentricidad
nx.radius(G2)
```

```
#periferio, el grupo de nodos que tienen la misma exentricidad al radio
nx.periphery(G2)
```

```
#centro, el grupo de nodos que tienen la misma exentricidad al radio
nx.center(G2)
```

###ejemplo simple

```
G=nx.karate_club_graph()#cargamos un nodo
G=nx.convert_node_labels_to_integers(G,first_label=1)
```

```
nx.average_shortest_path_length(G) #promedio de la distancia entre los alumnos
```

```
nx.radius(G) 
```

```
nx.diameter(G) #la distancia maxima entre 2 alumnos
```

```
nx.center(G) #alumnos que estan en el centro
```

```
nx.periphery(G) 
```
